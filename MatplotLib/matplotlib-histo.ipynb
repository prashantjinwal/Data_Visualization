```python
import matplotlib.pyplot as plt
import numpy as np 
import random 

ml_student_age = np.random.randint(18,45,(100))
py_student_age = np.random.randint(15,40,(100))

# print(ml_student_age)
# print(py_student_age)
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age , bins, rwidth=0.8, histtype="bar",orientation='horizontal', color="y", label= "ML Student")
# plt.hist(ml_student_age , bins, rwidth=0.8, histtype="bar")

plt.title("ML student age histograms")
plt.xlabel("Student age category")
plt.ylabel("Number of student age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (16,8))
plt.hist([ml_student_age, py_student_age] , bins, rwidth=0.8, histtype="bar", color=["m","y"], label= ["ML Student","PY Student"])
plt.title("ML student age histograms")
plt.xlabel("Student age category")
plt.ylabel("Number of student age")
plt.legend()
plt.show()
```
