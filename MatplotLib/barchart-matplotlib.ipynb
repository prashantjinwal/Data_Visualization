```python
df_Gplay = pd.read_csv("/kaggle/input/google-play-store-apps/googleplaystore.csv", nrows = 1000)
df_Gplay
```

```python
x = df_Gplay["Rating"]
y = df_Gplay["Reviews"]
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,8))
x = df_Gplay["Rating"]
y = df_Gplay["Reviews"]
plt.scatter(x,y, c = "r",marker="*", s=100, alpha=0.5, )#linewidth=4, edgecolor='g'
plt.scatter(x,df_Gplay["Installs"] , c = "m",marker="*", s=100, alpha=0.5, )
plt.title("Google Play Store Apps Scatter plot", fontsize = 20)
plt.xlabel("Rating", fontsize = 16)
plt.ylabel("Reviews & installs", fontsize = 16)
```
