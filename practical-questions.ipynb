```python
#  [ Q - 10 ]
# Write program in python
# a)	To store given dataset in pandas dataframe
# b)	Group the given dataset according to year of release
# c)	Apply data aggregation , agg() function to determine average movie rating per year.


data ={"S.no":[1,2,3,4,5,6],
      "Movies Name":["Movies1","Movies2","Movies3","Movies4","Movies5","Movies6"],
      "Year_of_release":[2032,2022,2012,2010,2025,2026],
      "Rating":[8.5,9.5,2.2,5.5,8.8,8.2],
      "Duration":[1.6,0.21,2.51,1.30,1.42,1.48]
     }

df = pd.DataFrame(data)
gp_df = df.groupby('Year_of_release')
for group_name, group_data in gp_df:
    print(f"Year: {group_name}")
    print(group_data)
    print("\n")
    
result = gp_df.agg({'Rating':'mean'})
print(result)
```

```python
# [ Q->12 ]
# a)	Write a python code to sort the given dataset according to Salary and display the name of employee having highest salary
# b)	What different types of plots can be used for visualization of categorical data in above dataframe.
# c)	Write a program code to display the line plot between salary vs age abd classified according to status.


data = {'Name':["John","Reeya","Linda","Vishal","Mohan","Alia"],
        'Age':[21,23,40,35,26,30],
        'Salary':[30000,40000,100000,75000,45000,60000],
        'Department':["sales","Marketing","Operations","Marketing","Finance","Finance"],
        'Status':["Junior","Junior","Senior","Senior","Junior","Senior"],
}

df = pd.DataFrame(data)
df_Sorted = df.sort_values(by='Salary', ascending=False)
highest = df_Sorted.iloc[0]["Name"]
print("the highest Salary : ", highest)

# [ii]
sns.lineplot(data=df, x="Salary", y="Name", markers=True, style='Status')
plt.title("Line plot Salary vs Name", fontsize=20)
plt.xlabel("Salary",fontsize=16)
plt.ylabel("Name", fontsize=16)
plt.show()
```
